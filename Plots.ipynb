```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
ruta = '/content/drive/My Drive/Crypto_2021/Code/sample_data/'
```

```python
data_ETH_close = pd.DataFrame(pd.read_csv(str(ruta)+'ETH-USD.csv'))['Close']
data_BTC_close = pd.DataFrame(pd.read_csv(str(ruta)+'BTC-USD.csv'))['Close']
dates = pd.DataFrame(pd.read_csv(str(ruta)+'BTC-USD.csv'))['Date']
```

```python
n = len(data_BTC_close)
x = np.arange(0,n)
```

```python
y_1 = data_ETH_close
y_2 = data_BTC_close 

fig, ax1 = plt.subplots()

labels = [0,'Sep. 2015', 'Jan. 2017', 'May 2019', 'Oct. 2019','Feb. 2021']

ax2 = ax1.twinx()

ax2.set_xticks([0,500,1000,1500,2000], minor=True)
ax2.set_xticklabels(labels, fontdict=None, minor=False, rotation=45)

ax1.plot(x, y_1, 'g-')
ax2.plot(x, y_2, 'b-')

ax1.set_title("Price evolution of ETH and BTC")
ax1.set_xlabel('Date')
ax1.set_ylabel('Ethereum/USD price', color='g')
ax2.set_ylabel('Bitcoin/USD price', color='b')

plt.show()
```
